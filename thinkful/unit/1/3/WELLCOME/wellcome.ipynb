```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
%matplotlib inline
pattern = re.compile('(\d{7,8})')
```

```python
def clean_cost(x):
    # remove the currency symbol
    try:
        x = re.sub('[^\d\.\,\s]+','',x)
        return float(x)
    except:
        return np.nan
    

def clean_title(x):
    title = x.strip('.').strip('\n')
    title = re.sub('\sj\s',' journal of ', title)
    title = title.replace('j ', 'journal of ')
    title = title.replace('immunol','immunology')
    title = title.replace('j.',"journal of ")
    title = title.replace('jnl', 'journal of ')
    title = title.replace('chem ','chemistry ')
    title = title.replace('chems', 'chemistry')
    title = title.replace('biochem','biochemistry')
    title = title.replace('am ','american ')
    title = title.replace('chembiochem', 'biochemistry')
    title = title.replace('trop med hyg', 'american journal of tropical medicine & hygiene')
    title = title.replace('nerv ', 'nervous ')
    title = title.replace('syst ', 'system ')
    title = title.replace('biol.', 'biology')
    title = title.replace('curr ', 'current ')
    title = title.replace('bioeth', 'bioethics')
    title = title.replace('int ', 'international ')
    title = title.replace('clin ', 'clinical ')
    title = title.replace('microbiol', 'microbiology')
    title = title.replace('cardiovasc magn reson', 'cardiovascular magnetic resonance')
    title = title.replace('infect', 'infectious')
    title = title.replace('dis ', 'diseases ')
    title = title.replace('mol ', 'molecular ')
    title = title.replace('radiat.', 'radiation')
    title = title.replace('pharm ', 'pharmaceutics ')
    pnas = 'proceddings of the national academy of sciences of usa'
    title = title.replace('pnas (proceedings of the national academy of sciences of the united states of america', pnas)
    title = title.replace('pnas usa', pnas)
    title = title.replace('proc natl acad sci u s a.', pnas)
    title = title.replace('proceedings of the national academy of sciences of the united states of america', pnas)
    title = title.replace('proceedings of the national academy of sciences of the united states of america', pnas)
    title = title.replace('proceedings of the national academy of sciences of the usa', pnas)
    title = title.strip()
    return title
    
def get_id(cost, length):
    try:
        cost = str(cost).strip('\n')
        iterator = pattern.finditer(cost)
        for match in iterator:
            # looking for length 7
            start, end = match.span()[0], match.span()[1]
            if (end - start) == length:
                return cost[start:end]
    except:
        return None
```

```python
# different encodings?
converters = {"COST (Â£) charged to Wellcome (inc VAT when charged)" : clean_cost, 
              "Journal title" : clean_title}
df = pd.read_csv("WELLCOME_APCspend2013_forThinkful.csv", encoding="latin1", converters=converters)
# rename our cols
df.columns = ['PMID/PMCID', 'Publisher', 'Journal title', 'Article title', 'Cost']
df["Journal title"] = df["Journal title"].apply(clean_title)
# drop missing for now
df.dropna(subset=['PMID/PMCID','Journal title', 'Cost'], inplace=True)
df["Journal title"] = df["Journal title"].apply(lambda x: x.lower() )
df["PMCID"] = df["PMID/PMCID"].apply(get_id, length = 7)
df["PMID"] = df["PMID/PMCID"].apply(get_id, length = 8)
df.drop(["PMID/PMCID"], axis=1, inplace=True)
# 999999 does not seem like a reasonable price. lets also drop these as they will 
# this price does not make sense...dramatically effect our stats
df = df.query("Cost != 999999.00")
```

```python
# how do we want to handle those records where it is missing both the PMCID and PMID?
mask = (df.PMCID.isna()) & (df.PMID.isna())
non_missing = df[~mask]
```
