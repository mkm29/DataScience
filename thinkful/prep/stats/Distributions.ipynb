```python
def plot(variable, y_max):
    plt.hist(variable)
    plt.vlines(np.mean(variable), 0, y_max, linestyles="dashed")
    plt.vlines(np.std(variable), 0, y_max)
    plt.legend(["mean", "standard deviation"])
    plt.show()
```
